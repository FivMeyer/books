# Introduo  Anlise Exploratria de Dados

No primeiro captulo o autor aborda rapidamente algumas aplicaes da estatstica e apresenta alguns exemplos cotidianos onde ela  usada. De modo geral a Estatstica est dividia em 3 grande reas, sendo elas:
* Estatstica Descritiva
* Probabilidade
* Inferncia Estatstica
<br/>

A **Estatstica Descritiva** , de certa forma, um conjunto de tcnicas destinadas a descrever e resumir os dados para que assim possamos tirar concluses a respeito de caractersticas de interesse.<br/>

A **Probabilidade** pode ser exemplificada como a teoria matemtica utilizada para se estudar a incerteza oriunda de fenmenos de carter aleatrio.<br/>

A **Inferncia Estatstica**  a area que estuda as tcnicas que possibilitam a extrapolao das informaes e concluses obtidas a partir de subconjunto de valores.<br/>

Ao conjunto de dados que contem a caracterstica de interesse  dado o nome de **Populao**. A populao contem todos os indviduos de um determinado grupo. Dessa forma podemos exemplificar a populao como:
1. Todos os habitantes de uma determinada cidade/estado/pas...;
2. Todas as mesas produzidas por uma fbrica de mveis;
3. Todo o sangue no corpo de uma pessoa...<br/>

No entanto, algumas vezes no conseguimos acessar todos os indivduos de uma populao sendo necessrio fazer o uso de uma pequena parcela desta para coletarmos os dados. Essa parcela  chamada de **Amostra**. A definio da parcela amostral  chamada de amostragem e ela tenta fornecer um subconjunto de valores o mais parecido possvel com a populao que lhe d origem.
